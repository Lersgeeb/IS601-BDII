## Creating a class that will make the different insert operations

```python
import random
from datetime import datetime, timedelta

```

```python
class sqlProjectCreator:
    def createGenreInsert(self, values):
        valuesStr = list(map(lambda x: "('%s')"%x, values ))
        insertStmt = 'INSERT INTO TBL_GENRE values %s ;' % (",".join(valuesStr))
        return insertStmt
    
    def createCategoryInsert(self, values):
        valuesStr = list(map(lambda x: "('%s')"%x, values ))
        insertStmt = 'INSERT INTO TBL_CATEGORY values %s ;' % (",".join(valuesStr))
        return insertStmt
    
    def createDevInsert(self, values):
        valuesStr = list(map(lambda x: "('%s')"%x, values ))
        insertStmt = 'INSERT INTO TBL_DEVELOPER values %s ;' % (",".join(valuesStr))
        return insertStmt


    def createGameInsert(self, games, publisherList):

        gameParseList = []
        for index, game in games.iterrows():
            name = game['Title']
            if '\'' in name: continue
            
            release_date = game['Release.Year']
            duration = int(game['Length.Main + Extras.Average'])
            if(duration == 0):  duration = 1
            dev_id = publisher.index(game['Metadata.Publishers']) + 1
            cat_id = game['Release.Rating']
            
            
    def createGameGenreInsert(self, myGameTable, gameSource):
        genreDict = {'Adventure':6,'Sgameports':9,'Sports':8,'Racing / Driving':2,'Action':3,'Educational':1,'Simulation':5,'Role-Playing (RPG)':4,'Strategy':7}    
        genreInsertValues = []
        
        for index, game in myGameTable.iterrows():
            name = game['GM_NAME']
            sourceRow = gameSource.loc[gameSource['Title'] == name]
            genreList = sourceRow['Metadata.Genres'].tolist()
            genreList = genreList[0].split(',')
            
            for genre in genreList:
                
                genreValue = "(%s, %s)" % (game['GM_ID'], genreDict[genre])
                genreInsertValues.append(genreValue)
            
        
        insertStmt = 'INSERT INTO TBL_GAME_GENRE VALUES %s ;' % (",".join(genreInsertValues))
        return insertStmt
            
    def CreateCatalogueInsert(self, myGameTable, catalogueList):
        catalogueInsertValues = []
        
        for index, game in myGameTable.iterrows():
            catalogueRow = catalogueList.loc[catalogueList['GM_ID'] == game['GM_ID']]
            catalogueDiscount = catalogueRow["DISCOUNT"]
            catalogueInsertValues.append('( %s, %s )' % (float(catalogueDiscount), int(game['GM_ID'])))
            
        insertStmt = 'INSERT INTO TBL_CATALOGUE VALUES %s ;' % (",".join(catalogueInsertValues))
        return insertStmt
    
    def CreateUserInsert(self, userTable):
        userInsertValues = []
        
        for index, user in userTable.iterrows():
            username = user['USERNAME']
            email = user['EMAIL']
            pw = user['PASSWORD'] 
            playing = user['IS_PLAYING']
            
            userValue = "\n\t( '%s', '%s', '%s', %s )" % (username, email, pw, int(playing))
            
            userInsertValues.append(userValue)
            
        insertStmt = 'INSERT INTO TBL_USER VALUES %s \n;' % (",".join(userInsertValues))
        return insertStmt
    
    def CreateSaleInsert(self, userTable, catalogueTable):
        
        userRandomSale = []
        count = 0
        for index, user in userTable.iterrows():
            #Crear una lista de compras totales por usuario mediante un numero random entre 0 - 10
            totalSales = random.randint(0, 10)
            
            for sale in range(totalSales):
                 #Seleccionar juegos random 
                randomGame = random.randint(0, catalogueTable.shape[0] - 1)
                rndCatalogueId = catalogueTable.iloc[randomGame]["CATLG_ID"]
                

                #15% de los juegos comprados que sean regalos
                isGift = False
                randomInt = random.randint(1, 20)
                if(randomInt <= 3):
                    isGift = True

                #fechas random pero si son regalos intentar que en un 60% sean el 14/02 o 24/12
                
                inicio = datetime(2013, 1, 1)
                final =  datetime(2020, 12, 31)
                ranDate = inicio + (final - inicio) * random.random()
                
                if(isGift):
                    randomInt = random.randint(1, 20)
                    if(randomInt <= 12):
                        year = random.randint(2013, 2020)
                        day = 24
                        month = 12
                        if(randomInt <=6):
                            day = 14
                            month = 2
                            
                        ranDate = datetime(year, month, day)
                    
                    
                dateStr = '%s/%s/%s' % (ranDate.year, ranDate.month, ranDate.day)
                insertValue = " (%s, %s, '%s', %s)" % (rndCatalogueId, user["USER_ID"], dateStr, int(isGift))    
                if(count%3 == 0):
                    insertValue = "\n\t%s" % insertValue
                count += 1
                userRandomSale.append(insertValue)
            
        insertStmt = 'INSERT INTO TBL_SALE VALUES %s \n;' % (",".join(userRandomSale))
        return insertStmt
    
    def createGiftInsert(self, saleTable, userTable):
        giftInsertValues = []
        giftDict = {}
        count = 0
        for index, sale in saleTable.iterrows():
            if(sale['IS_GIFT']):
                rndUserId = sale['USER_ID']
                while(rndUserId == sale['USER_ID']):
                    randomUser = random.randint(0, userTable.shape[0] - 1)
                    rndUserId = userTable.iloc[randomUser]["USER_ID"]
                
                
                insertValue = "(%s, %s)" % (sale['SALE_ID'], rndUserId)
                giftDict[sale['SALE_ID']] = rndUserId
                if(count%6 == 0):
                    insertValue = "\n\t%s" % insertValue
                giftInsertValues.append(insertValue)
                count += 1
                
        insertStmt = 'INSERT INTO GAME_GIFT VALUES %s \n;' % (",".join(giftInsertValues))
        
        return (insertStmt, giftDict)
            
    def createOwnerInsert(self, saleTable, giftDict, catalogTable):
        ownerInsertValues = []
        count = 0
        for index, sale in saleTable.iterrows():
            ownerId = sale['USER_ID']
            if(sale['IS_GIFT']):
                ownerId = int(giftDict[sale['SALE_ID']])
                
            catalogueRow =  catalogTable[catalogTable['CATLG_ID'] == sale['CATLG_ID']]
            insertValue = "(%s, %s)" % (ownerId, int(catalogueRow['GM_ID']))
            if(count%6 == 0):
                insertValue = "\n\t%s" % insertValue
            ownerInsertValues.append(insertValue)                
            count += 1
            
            
                
        insertStmt = 'INSERT INTO TBL_GAME_OWNER VALUES %s \n;' % (",".join(ownerInsertValues))
        
        return insertStmt
    
    def CreateSessionInsert(self, gameOwner, myGameTable, saleTable, gifts, catalogue):
        
        game_session_values = []
        count = 0
        for index, game_played in gameOwner.iterrows():
            
            #Fecha cuando se obtuvo el juego
            currentGameId = game_played['GM_ID'] 
            catalogueRow = catalogue[catalogue['GM_ID'] == currentGameId]
            catalogueId =int( catalogueRow['CATLG_ID'])
            saleRows = saleTable[saleTable['CATLG_ID'] == catalogueId]
            rowExist = saleRows[saleRows['USER_ID'] == game_played['USER_ID']]


            #OBTENER VENTA DEL JUEGO  Y fECHA DE COMPRA
            if (not rowExist.empty and rowExist.empty == False ):
                saleRowFound = rowExist
            else:
                for index, saleRow in saleRows.iterrows():
                    saleId = saleRow['SALE_ID']
                    if saleId in gifts:
                        if (gifts[saleId] == game_played['USER_ID']):
                            saleRowFound = saleRow
                    else:
                        continue
                    
            #session inicial 
            fecha_sql = saleRowFound['SALE_DATE']
            fechaStr = fecha_sql                
            if(not type(fecha_sql) == str):
                fecha_sql = fecha_sql.tolist()
                fechaStr = fecha_sql[0]
                
            
            fecha_inicio = datetime.strptime(fechaStr, '%Y-%m-%d %H:%M:%S')
            fecha_final =  datetime(2021, 3, 25)                
            ranDate = fecha_inicio + (fecha_final - fecha_inicio) * random.random()    
            dateStr = '%s/%s/%s' % (ranDate.year, ranDate.month, ranDate.day)
                
            
            #Crear una lista de sesiones totales por juego del usuario mediante un numero random entre 0 - 10
            prox_inicio = ranDate
            totalSessions = random.randint(0, 3)
            for session in range(totalSessions):
                inicio = prox_inicio
                hours = 8
                hours_added = timedelta(hours = hours)
                fecha_final_limite =  inicio + hours_added
                final = inicio + (fecha_final_limite - inicio) * random.random()    
                delta = final - inicio
                horasDeJuego = (round(delta.total_seconds() / 3600.0 ,2))
                
                
                insertValue = "(%s, %s, %s, '%s', '%s')" % (
                    game_played['USER_ID'],
                    game_played['GM_ID'],
                    horasDeJuego ,
                    inicio.strftime("%Y/%m/%d %H:%M:%S"),
                    final.strftime("%Y/%m/%d %H:%M:%S"),
                )
                 
                if(count%3 == 0):
                    insertValue = "\n\t%s" % insertValue
                
                game_session_values.append(insertValue)
                count += 1
                
                
    
        insertStmt = 'INSERT INTO TBL_GAME_SESSION VALUES %s \n;' % (",".join(game_session_values))
        return insertStmt
        
        
#instance of the class
        
sqlPro = sqlProjectCreator()
```
